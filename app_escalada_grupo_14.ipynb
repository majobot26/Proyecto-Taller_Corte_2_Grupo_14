## Descripción
Este proyecto implementa un sistema de gestión de clientes, productos y pedidos utilizando Programación Orientada a Objetos (POO) y una base de datos SQLite.

## Programación Orientada a Objetos

Se implementan las siguientes clases:

- **Persona (Clase Padre)**
- **Cliente (Clase Hija)**
- **Producto**
- **Pedido**

Se aplican conceptos de:
- Herencia
- Encapsulamiento
- Métodos (get/set implícitos)

```python
import sqlite3

conn = sqlite3.connect("pasteleria.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS clientes (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT,
    telefono TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS productos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT,
    precio REAL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS pedidos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    cliente_id INTEGER,
    producto_id INTEGER,
    cantidad INTEGER,
    FOREIGN KEY(cliente_id) REFERENCES clientes(id),
    FOREIGN KEY(producto_id) REFERENCES productos(id)
)
""")

conn.commit()
```

```python
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

class Cliente(Persona):
    def __init__(self, nombre, telefono):
        super().__init__(nombre)
        self.telefono = telefono

    def guardar(self):
        try:
            cursor.execute("INSERT INTO clientes (nombre, telefono) VALUES (?, ?)", 
                           (self.nombre, self.telefono))
            conn.commit()
        except Exception as e:
            print("Error:", e)

class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def guardar(self):
        try:
            cursor.execute("INSERT INTO productos (nombre, precio) VALUES (?, ?)", 
                           (self.nombre, self.precio))
            conn.commit()
        except Exception as e:
            print("Error:", e)

class Pedido:
    def __init__(self, cliente_id, producto_id, cantidad):
        self.cliente_id = cliente_id
        self.producto_id = producto_id
        self.cantidad = cantidad

    def guardar(self):
        try:
            cursor.execute("""
                INSERT INTO pedidos (cliente_id, producto_id, cantidad)
                VALUES (?, ?, ?)
            """, (self.cliente_id, self.producto_id, self.cantidad))
            conn.commit()
        except Exception as e:
            print("Error:", e)
```

```python
def ver_clientes():
    cursor.execute("SELECT * FROM clientes")
    return cursor.fetchall()

def actualizar_cliente(id, nuevo_nombre):
    cursor.execute("UPDATE clientes SET nombre=? WHERE id=?", (nuevo_nombre, id))
    conn.commit()

def eliminar_cliente(id):
    cursor.execute("DELETE FROM clientes WHERE id=?", (id,))
    conn.commit()

def ver_pedidos_completos():
    cursor.execute("""
        SELECT clientes.nombre, productos.nombre, pedidos.cantidad
        FROM pedidos
        JOIN clientes ON pedidos.cliente_id = clientes.id
        JOIN productos ON pedidos.producto_id = productos.id
    """)
    return cursor.fetchall()
```

```python
# Clientes
Cliente("Maria", "123").guardar()
Cliente("Ana", "456").guardar()
Cliente("Luisa", "789").guardar()

# Productos
Producto("Torta Chocolate", 50000).guardar()
Producto("Cheesecake", 45000).guardar()
Producto("Cupcake", 8000).guardar()

# Pedidos
Pedido(1, 1, 2).guardar()
Pedido(2, 2, 1).guardar()
Pedido(3, 3, 5).guardar()
```

```python
print("Clientes:")
print(ver_clientes())

print("\nPedidos (JOIN):")
print(ver_pedidos_completos())

conn.close()
```
